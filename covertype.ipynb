# Predicting forest cover type in Northern Colorado

![alt text](https://www.dropbox.com/s/qq9kp4v6fkx90ec/colorado-forest-cover.png?dl=1)

This notebook is meant as a tutorial to the [Kaggle Forest Cover Type Prediction](http://www.kaggle.com/c/forest-cover-type-prediction) challenge. 

The goal of the competition is to predict the predominant tree cover from strictly cartographic variables (as opposed to remotely sensed data). The actual forest cover type for a given 30 x 30 meter cell was determined from US Forest Service (USFS) Region 2 Resource Information System data. Independent variables were then derived from data obtained from the US Geological Survey and USFS. The data is in raw form (not scaled) and contains binary columns of data for qualitative independent variables such as wilderness areas and soil type.

This study area includes four wilderness areas located in the Roosevelt National Forest of northern Colorado. These areas represent forests with minimal human-caused disturbances, so that existing forest cover types are more a result of ecological processes rather than forest management practices.

Data is provided by Bache, K. & Lichman, M. (2013). UCI Machine Learning Repository. Irvine, CA: University of California, School of Information and Computer Science.

## Predictive Modeling

### Decision Trees

Binary tree of feature threshold functions that recursively partitions the input space. Trees are built in a greedy fashion to minimize the impurity of partitions (e.g. a partition should only contain areas of the same cover type).

![Decision tree](https://www.dropbox.com/s/o3zh7djv21034ah/tree.png?dl=1)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    fig = plt.figure(figsize=(8, 6))
    ax = fig.gca()
    cs = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.set_title(title)
    plt.colorbar(cs)
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names, rotation=45)
    plt.yticks(tick_marks, class_names)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


def evaluate_dtree(feature_names, **kwargs):
    X = df[feature_names].values.astype(np.float32)
    y = df['Cover_Type'].values

    X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=1234, test_size=0.2)

    cart = DecisionTreeClassifier(min_samples_leaf=5, **kwargs)

    cart.fit(X_train, y_train)
    print('ACC: %.4f' % cart.score(X_val, y_val))
    cm = confusion_matrix(y_val, cart.predict(X_val))
    plot_confusion_matrix(cm)
    return cart
```

```python
feature_names = ['Elevation']
cart = evaluate_dtree(feature_names, max_depth=2)
```
